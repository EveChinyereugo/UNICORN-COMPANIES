Question 1- Which unicorn companies have had the biggest Return On Investment?

```python
# taking out the  "$", "B", "M" from Funding column 

Unicorn_Companies.Funding = Unicorn_Companies.Funding.str.replace("$","")

Unicorn_Companies.Funding = Unicorn_Companies.Funding.str.replace("B",'0'*9)

Unicorn_Companies.Funding = Unicorn_Companies.Funding.str.replace("M",'0'*6)
```

```python
import pandas as pd

# Assuming you have a DataFrame called 'Unicorn_Companies' and you want to convert the 'Funding' column to float
# Replace 'Unknown' with NaN in the 'Funding' column
Unicorn_Companies['Funding'] = pd.to_numeric(Unicorn_Companies['Funding'], errors='coerce')

# Now you can convert the column to float
Unicorn_Companies['Funding'] = Unicorn_Companies['Funding'].astype(float)
```

```python
# Convert the 'Funding' column to float
Unicorn_Companies['Funding'] = Unicorn_Companies['Funding'].astype(float)
```

```python
# taking out the  "$", "B", "M" from Valuation column 
Unicorn_Companies['Valuation'] = Unicorn_Companies['Valuation'].astype(str).str.replace('$', '').str.replace('B', '0'*9).str.replace('M', '0'*6)
```

```python
# convert the 'Valuation' column to float
Unicorn_Companies['Valuation'] = Unicorn_Companies['Valuation'].astype(float)
```

```python
# Calculate ROI for each company

Unicorn_Companies['ROI'] = Unicorn_Companies['Valuation'] - Unicorn_Companies['Funding']
 
# Sort the dataframe by ROI in descending order and select the top 10 companies

sorted_df = Unicorn_Companies.sort_values('ROI', ascending=False)

 
# Display the top 10 companies with the biggest ROI

print(sorted_df[['Company', 'ROI']])

sorted_df = Unicorn_Companies.sort_values('ROI', ascending=True).head(10)
```

```python
# calculate ROI for each company

Unicorn_Companies['ROI'] = Unicorn_Companies['Valuation'] - Unicorn_Companies['Funding']

# Sort the dataframe by ROI in descending order and select the top 10 companies

sorted_df = Unicorn_Companies.sort_values('ROI', ascending=False).head(10)

 
# Create a horizontal bar plot

plt.figure(figsize=(10, 6))

plt.barh(sorted_df['Company'], sorted_df['ROI'])

plt.xlabel('Return on Investment (in billions)')

plt.ylabel('Company')

plt.title('Top 10 Unicorn Companies with the Biggest Return on Investment')

plt.show()
```

Question 2 -How long does it usually take for a company to become a unicorn? Has it always been this way?

```python
# We need to find the lenght of time it takes for a company to be a unicorn
# first we need to Convert the JoinDate column to datetime


Unicorn_Companies["Date Joined"] = pd.to_datetime(Unicorn_Companies["Date Joined"])

 
# then crete a `YearsToJoin` column.

 
Unicorn_Companies["Years To Join"] = Unicorn_Companies["Date Joined"].dt.year - Unicorn_Companies["Year Founded"]


# Display the first few rows of `companies`to confirm that the new column did get added.
Unicorn_Companies.head()
```

```python
Unicorn_Companies['Years To Join'].value_counts()
```

```python
import matplotlib.pyplot as plt

 
grouped_df = Unicorn_Companies.groupby('Year Founded')['Years To Join'].mean()

plt.plot(grouped_df.index, grouped_df.values)
plt.xlabel('Year Founded')
plt.ylabel('Average Years to Join')
plt.title('Average Years to Join for Unicorn Companies by Year Founded')
plt.show()
```

## Question - 4. Which Investors have funded the most unicorns?


```python
# Investor that funded the most unicorn companies

# Calculate Select Investor by their Company funding.

top_investors = Unicorn_Companies['Select Investors'].value_counts().head(5).index.tolist()

print(f"The top five select investors who funded unicorn companies are: {top_investors}")
```

```python
import matplotlib.pyplot as plt

# Get the data for the top five select investors
top_investors = Unicorn_Companies['Select Investors'].value_counts().head(5).index.tolist()
investor_data = Unicorn_Companies[Unicorn_Companies['Select Investors'].isin(top_investors)].groupby('Select Investors')['Funding'].sum().sort_values(ascending=False)

# Create a donut chart of the investor data
plt.figure(figsize=(8, 8))
plt.pie(investor_data.values, labels=investor_data.index, autopct='%1.1f%%', startangle=90, pctdistance=0.85, labeldistance=1.05)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Total Funding by Select Investor (Top 5)')
plt.show()
```

```python
# filter the unicorn_data dataframe to include only rows where the 'Select Investors' column is null
null_investors = Unicorn_Companies[Unicorn_Companies['Select Investors'].isnull()]

# print the resulting dataframe
(null_investors)
```

```python
# Sort the dataframe by Select Investors in descending order and select the top 10 companies
sorted_df = Unicorn_Companies.sort_values('Funding', ascending=False).head(10)


# Create a horizontal bar plot
plt.figure(figsize=(10, 6))
plt.barh(sorted_df['Select Investors'], sorted_df['Funding'])
plt.xlabel('Investors with the most funding (in billions)')
plt.ylabel('Select Invesor')
plt.title('Top 10 Select Investors and their funding')
plt.show()
```
